```python
from transformers import pipeline

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

ticket = "I was charged twice on my credit card."

labels = [
    "Billing Issue",
    "Technical Problem",
    "Refund Request",
    "Account Access",
    "General Inquiry"
]

result = classifier(ticket, labels)
```

Get Top 3 Tags

```python
top3 = sorted(zip(result["labels"], result["scores"]),
              key=lambda x: x[1],
              reverse=True)[:3]

print("Top 3 Tags:", top3)
```
